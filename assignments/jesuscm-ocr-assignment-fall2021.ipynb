```python
## Results are not at all clean.  Some files are text-readable and some are just images.
## The easiest to convert to image and then use ocr before using regex

## Let's test on first file using the code from class

PDF_file = cal_rev_files[0]

print(f"Converting '{PDF_file}' to a JPEG")
pages = convert_from_path(PDF_file, 300) ## could opt for different dpis. balance between hi-res and computational power

image_counter = 1

for page in pages:
    print(f"the page type is {type(page)}")
    ## Declaring filename for each page of PDF as JPG
    filename = str(PDF_file[ : -4])+"_page_"+str(image_counter)+".jpg"  ## remove the .pdf extention (last 4 characters)
    print(f"Here we go: {filename}")
    # Save the image of the page in system
    page.save(filename, 'JPEG') ## here we tell it to be a "jpeg" file
    # Increment the counter to update filename
    image_counter+=1
```

```python
## this is the first attempt at the regex needed to extract the date from the filename for better organization

date_from_filename =  "(\w{3,5})[-_]\d{0,2}(\d{2})"
```

```python
## this is a test loop to convert to images on first four files
## we'll use date_from_filename to search through those files and created jpg images
## with more standardized names.

jpgs_dir = Path(f"{base_path}ocr-jpgs")  ## to stay organized, put files in directory
jpgs_dir.mkdir(exist_ok=True)  ## make ocr-jpgs directory

for file in cal_rev_files[0:4]:
    ## convert from path takes a pdf and turns it into a jpg
    pages = convert_from_path(file, 300)  ## trying it at the default 300dpi
    page_counter = 1
    for page in pages:
        ## extract month and year from original pdf file name in order to standardize it
        name_components = re.findall(date_from_filename, file)[0]
        ## jpg file names are this format: "year-month-finance-bulletin-pagenum"
        filename = f"{jpgs_dir}/{name_components[1]}-{name_components[0][0:3]}-finance-bulletin-p{page_counter}.jpg"
        page.save(filename, 'JPEG')
        page_counter += 1
```

```python
## this worked for converting them to jpegs.  Let's turn it into a function:

jpgs_dir = Path(f"{base_path}ocr-jpgs")  ## to stay organized, put files in director
file_names = []  ## for holding the standardized filenames for later usage
file_paths = []  ## for holding the path files to the jpg images for later usage

def to_jpg(file, resolution = 300):
    """
    This function turns pdfs into jpgs, and saves each page as a different jpg file.  The arguments it takes are
    the location of the pdf file and the resolution.  By default, the resolution is set to 300.
    """
    
    print("Getting PDF from the target folder.  Please wait...")
    
    jpgs_dir.mkdir(exist_ok=True)  ## make ocr-jpgs dir

    pages = convert_from_path(file, resolution)  ## trying it at 300dpi
    page_counter = 1
    for page in pages:
        ## extract march and year from original pdf file name in order to standardize it
        name_components = re.findall(date_from_filename, file)[0]
        ## jpg file names are this format: "year-month-finance-bulletin-pagenum"
        file_name = f"{name_components[1]}-{name_components[0][0:3]}-finance-bulletin-p{page_counter}.jpg"
        file_path = f"{jpgs_dir}/{file_name}"
        page.save(file_path, 'JPEG')
        file_names.append(file_name)
        file_paths.append(file_path)
        page_counter += 1
```

```python
## let's test this out on the first four files too

for file in cal_rev_files[ :4]:
    to_jpg(file)
    print(file_names)
    print(file_paths)
```

```python
## the function worked.  Now let's try a function to OCR on those first four files 
## and turn them into txt files.  The name of the txt files will be the same as the jpg
## files but .txt instead of .jpg

outfiles = []  ## this will hold the paths to the txt files that are created by this function.  This is more global
                ## than the function so they persist for later usage

def jpg_to_txt (source_img, source_path):
    '''
    To convert a JPG to text -- arguments are: the name of the jpg file, and its path.
    '''    
    
    output_path = Path(f"{base_path}output/")  ## where the txt files will be saved
    output_path.mkdir(exist_ok = True)  ## make the output directory in case it doesn't yet exist
    outfile = os.path.join(output_path,f"{source_img[: 23]}.txt")  ## this will be the path fo the actual file
    outfiles.append(outfile)  ## append it to the outfiles list so it can be referenced later
    
    with open(outfile, "w") as txtfile:  ## I have changed this from "a" to "w" so it doesn't append the same data to already existing files
        print(f"Converting image {source_img} to a text file!")
        text = f"\n\n\nFile Info: {source_img} \n\n\n \
        {str(pytesseract.image_to_string(Image.open(source_path)))}"
        
        text = text.replace("-\n", "")
        txtfile.write(text)
        
        ##shutil.rmtree(source_path)
        print(f"A text file named {outfile} is ready for you")
```

```python
for file_name in file_names:
    jpg_to_txt(file_name, file_paths[file_names.index(file_name)])
```

```python
## jpg_to_txt function was a success!

## now let's create a function to reset the files that are stored in global variables
## reset_files will be called whenever a new ocr process is called

def reset_files():
    file_names.clear()
    file_paths.clear()
    outfiles.clear()
```

```python
## combine both functions into an ocr_files function that takes in a list of files

def ocr_files(PDF_files, resolution = 300):
    """
    args:
        1.  A list of PDF-files in their directories
        2.  Resolution
    """
    reset_files()
    for file in PDF_files:
        to_jpg(file, resolution)
    for file_name in file_names:
        jpg_to_txt(file_name, file_paths[file_names.index(file_name)])
```

```python
## try ocr_files on the first four files
ocr_files(cal_rev_files[ :4], 300)
```

```python
## Success!!!

## now let's try to see how well the ocr worked on those four files using regex
## to get the info we're looking for at the default 300 resolution

## these are the global regex patterns we will use for testing as well as for the whole
## directory of pdf files
mon_year_pat = re.compile("^(?P<mon_year>[JFMASOND]\w*\s\d{4})", re.M)
alc_nos_pat = re.compile("Alcoholic Beverages\s(?P<forecast>\S{1,2})\s{1}(?P<actual>\S{1,2})")

for file in outfiles:
    with open(file, "r") as text_obj:
        text = text_obj.read()
        mon_year = re.search(mon_year_pat, text)
        print(mon_year.group("mon_year"))
        alc_stats = re.search(alc_nos_pat,text)
        print(alc_stats.group("forecast"))
        print(alc_stats.group("actual"))
```

```python
## That worked to get a pretty good ocr of the jpg files.
## Now that they're converted to text files, this function will now go through and
## run regex on the txt files in order to pull out the relevant information, load it
## into a database, and return that database...

def regex_text_files(files):
    mons_years = []
    forecasts = []
    actuals = []
    
    for file in files:
        with open(file, "r") as text_obj:
            text = text_obj.read()
            mons_years.append(re.search(mon_year_pat, text).group("mon_year").replace("\n", " ").capitalize())
            forecast = pd.to_numeric(re.search(alc_nos_pat, text).group("forecast"), errors = "coerce")  ## this forces NAN into the spaces where clear numbers weren't read
            forecasts.append(forecast*10e6)  ## to make them millions
            actual = pd.to_numeric(re.search(alc_nos_pat, text).group("actual"), errors = "coerce")
            actuals.append(actual*10e6)
    df = pd.DataFrame(zip(mons_years, pd.to_numeric(forecasts), pd.to_numeric(actuals)), columns = ["Date", "Forecast", "Actual"])
    return df
```

```python
## This is not a perfect data set.  Where there were problems, some hand-coding and/or
## other methods may be needed.  A good start to get the bulk of the information though.

alc_df = regex_text_files(outfiles)
alc_df.to_csv("alcoholic_drinks-forecast-actual.csv", encoding = "UTF-8", index = False)
```
